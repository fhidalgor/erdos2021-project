```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("whitegrid")
```

```python
bert_data = pd.read_csv("LearningRate4_TwoAbbr_BERT.txt")
electra_data = pd.read_csv("LearningRate4_TwoAbbr_Electra.txt")
```

```python
plt.scatter(bert_data["Epoch"], bert_data["ValidAcc"], label= "Validation Accuracy")
plt.scatter(bert_data["Epoch"], bert_data["TrainAcc"], label = "Train Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("BERT Train versus Validation ")
plt.legend()
# Validation accuracy is checked at the end of the training epoch.
```

```python
plt.scatter(bert_data["Epoch"], bert_data["ValidAcc"], label= "BERT")
plt.scatter(electra_data["Epoch"], electra_data["ValidAcc"], label = "Electra")
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.title("Accuracy Comparison")
plt.legend()
```

```python
plt.scatter(bert_data["Epoch"], bert_data["ValidLoss"], label= "BERT")
plt.scatter(electra_data["Epoch"], electra_data["ValidLoss"], label = "Electra")
plt.xlabel("Epoch")
plt.ylabel("Validation Loss")
plt.title("Loss Comparison")
plt.legend()
```
