```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("whitegrid")
```

```python
data1 = pd.read_csv("ElectraSmall_EightAbbr.txt")
data2 = pd.read_csv("ElectraSmall_EightAbbr_OldLoc.txt")
data3 = pd.read_csv("ElectraSmall_TwoAbbr_LR4.txt")
data4 = pd.read_csv("ElectraSmall_TwoAbbr_OldLoc.txt")
```

```python
plt.figure(figsize = (10,7))
plt.title("Accuracy Comparison")
plt.scatter(data1["Epoch"], data1["ValidAcc"], label= "Token 8Abbr", marker = "o", color = "blue")
plt.scatter(data1["Epoch"], data2["ValidAcc"], label= "MeDAL 8Abbr", marker = "o", color = "red")
plt.scatter(data1["Epoch"], data3["ValidAcc"][:10], label= "Token 2Abbr", marker = "x", color = "C0")
plt.scatter(data1["Epoch"], data4["ValidAcc"][:10], label= "MeDAL 2Abbr", marker = "x", color = 'orange')
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.legend()
```

```python
plt.figure(figsize = (10,7))
plt.title("Loss Comparison")
plt.scatter(data1["Epoch"], data1["ValidLoss"], label= "Token 8Abbr", marker = "o", color = "blue")
plt.scatter(data1["Epoch"], data2["ValidLoss"], label= "MeDAL 8Abbr", marker = "o", color = "red")
plt.xlabel("Epoch")
plt.ylabel("Validation Loss")
plt.legend()
```
