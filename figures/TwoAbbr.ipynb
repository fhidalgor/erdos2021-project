```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("whitegrid")
```

```python
data1 = pd.read_csv("ElectraSmall_TwoAbbr_LR4.txt")
data2 = pd.read_csv("ElectraSmall_TwoAbbr_OldLoc.txt")
data3 = pd.read_csv("ElectraSmall_TwoAbbr_CLS.txt")
```

```python
plt.figure(figsize = (10,7))
plt.title("Accuracy Comparison")
plt.scatter(data1["Epoch"], data1["ValidAcc"], label= "Token Locations", marker = "x")
plt.scatter(data1["Epoch"], data2["ValidAcc"], label= "MeDAL Locations", marker = "x")
plt.scatter(data1["Epoch"], data3["ValidAcc"], label= "CLS Locations", marker = "x")
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.legend()
```

```python
plt.figure(figsize = (10,7))
plt.title("Loss Comparison")
plt.scatter(data1["Epoch"], data1["ValidLoss"], label= "Token Locations", marker = "x")
plt.scatter(data1["Epoch"], data2["ValidLoss"], label= "MeDAL Locations", marker = "x")
plt.scatter(data1["Epoch"], data3["ValidLoss"], label= "CLS Locations", marker = "x")
plt.xlabel("Epoch")
plt.ylabel("Validation Loss")
plt.legend()
```
